```python
df = pd.read_csv("../Downloads/DTSense_Week_3/raw_data/sales_subset.csv", index_col = 0)
df.head(2)
```

```python
df.info()
```

```python
df.drop("unemployment", axis = 1, inplace = True)
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df["month"] = df["date"].dt.month
df.head(2)
```

```python
df["month_name"] = df["date"].dt.month_name()
df.head(2)
```

```python
#0 --> Monday, 6 --> Sunday
df["day_of_week"] = df["date"].dt.weekday
df.head(2)
```

```python
df["day_of_week_text"] = df["date"].dt.day_name()
df.head(2)
```

```python
df["day_of_month"] = df["date"].dt.day
df.head(2)
```

```python
df["week_of_year"] = df["date"].dt.isocalendar().week
df.head(2)
```

```python
df["day_of_year"] = df["date"].dt.day_of_year
df.head(2)
```

```python
df["year"] = df["date"].dt.year
df.head(2)
```

```python
# mengecek record data tersebut dari tahun berapa saja
df["year"].unique()
```

```python
# mengecek frequency data tersebut dari tahun berapa saja
df["year"].value_counts()
```

```python
from datetime import date
```

```python
df["current_date"] = pd.to_datetime(date.today())
```

```python
df.head(2)
```

```python
today_date = date.today()
df["day_to_today"] = today_date - df["date"].dt.date
df.head(2)
```

```python
# mengkonversi hitungan dari hari menjad hitungan tahun
df["year_to_today"] = ((df["current_date"] - df["date"]) // np.timedelta64(1, 'W'))
df.head(2)
```

```python
# Kalkulasi jumlah penjualan berdasarkan store-nya
df_stroe_sales = df.groupby("store", as_index = False)["weekly_sales"].sum()
df_stroe_sales
```
